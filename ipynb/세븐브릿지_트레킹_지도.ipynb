# 세븐브릿지, 트래킹, 해수욕장 좌표 dataframe 화

```python
#세븐브릿지 좌표
df_bridge = pd.read_csv('./files/좌표/세븐브릿지_좌표.txt')
df_bridge.columns=['브릿지','위도','경도']
df_bridge['위도']

#트래킹 좌표
df_mountain = pd.read_csv('./files/좌표/5대트래킹_좌표.txt')
df_mountain.columns=['트래킹','위도','경도']

# 해수욕장좌표
df_beach = pd.read_excel('./files/해수욕장좌표.xlsx')
df_beach
```

# 세븐브릿지 칼럼추가

```python
df_bridge['시군구'] = ['강서구','강서구','사하구','영도구','영도구','남구','수영구']
df_bridge['주변관광지']=['한국마사회 렛츠런파크 부산경남',
                       '한국마사회 렛츠런파크 부산경남',
                       '감천문화마을,부네치아 선셋 전망대',
                       '태종대 관광지',
                       '태종대 관광지',
                       '부산시립박물관',
                       '광안리해수욕장']
df_bridge                       
```

# 트래킹 칼럼 추가

```python
df_mountain['소요시간']=['2시간 30분', '2시간', '1시간 40분', '1시간 30분', '3시간 30분']
df_mountain['난이도']=['✮✮', '✮✮', '✮✮', '✮', '✮✮✮✮']
df_mountain['인증지']=['가덕도 연대봉', '봉래산 정상석', '광안대교뷰 or 오륙도가 보이는 뷰',\
                    '광안대교뷰 or 블루라인파크 열차가 보이는 뷰', '금정산 정상석']
df_mountain
```

# 지도 시각화

```python
file_path = "./maps/부산_시군구.geojson"
busan_sgg_get = json.load(open(file_path, encoding="utf-8"))
busan_sgg_get


print(len(busan_sgg_get['features']))
      
for i in range (len(busan_sgg_get['features'])):
    result = busan_sgg_get['features'][i]['properties']
    print(result)

m = folium.Map(
    location=[35.159144, 129.048293],
    #tiles = 'Stamen Terrain',
    zoom_start=10
    
)
    
tooltip = 'Click!'


#지도에 세븐브릿지 표시
for i in df_bridge.index:
    
    folium.Marker([df_bridge.loc[i, "위도"], df_bridge.loc[i, "경도"]],
                popup=f'<div style="width:200px"><strong>{df_bridge.iloc[i]["브릿지"]}</strong><br>\
                시군구 : {df_bridge.iloc[i]["시군구"]}<br>\
                주변관광지 : {df_bridge.iloc[i]["주변관광지"]}<br>\
                <img width="200px" src = ./image/{i}_bridge.jpg><br>\
                <a href="https://www.youtube.com/watch?v=7PZkgZrHKfI">YOUTUBE 홍보영상</a></div>',
                icon = folium.Icon(color='blue', icon='ok-circle'),
                
                tooltip = tooltip).add_to(m)
    



#지도에 트래킹 표시
for j in range(len(df_mountain)):
    
    folium.Marker([df_mountain.loc[j, "위도"],df_mountain.loc[j, "경도"]],
                 popup=f'<div style="width:200px"><strong>{df_mountain.iloc[j]["트래킹"]}</strong><br>\
                 소요시간 : {df_mountain.iloc[j]["소요시간"]}<br>\
                 난이도 : {df_mountain.iloc[j]["난이도"]}<br>\
                 인증지 : {df_mountain.iloc[j]["인증지"]}<br>\
                 <img width="200px" src = ./image/{j}_mountain.jpg><br>\
                 <a href="https://blog.naver.com/PostView.naver?blogId=busanto1115&logNo=222772366380&categoryNo=65&parentCategoryNo=&from=thumbnailList">관련 블로그</a></div>',
                 icon = folium.Icon(color='green', icon='ok-circle'),
              
                 tooltip = tooltip
                 ).add_to(m)

    
#지도에 해수욕장 표시
for i in df_bridge.index:
    
    folium.Marker([df_beach.loc[i, "위도"], df_beach.loc[i, "경도"]],
                popup=f'<div style="width:200px"><strong>{df_beach.iloc[i]["해수욕장"]}</strong><br>\
                <img width="200px" src = ./image/{i}_beach.jpg><br>\
                <a href="https://www.youtube.com/watch?v=xLD8oWRmlAE">YOUTUBE 홍보영상</a></div>',
                icon = folium.Icon(color='red', icon='ok-circle'),
                
                tooltip = tooltip).add_to(m)    


# style 지정
def style_function(feature) :
    df = {
        "opacity" : 0.5,
        "weight" : 0.6,
        "color" : "yellow",
        "fillOpacity" : 0
    }

    return df
    
folium.GeoJson(
    busan_sgg_get,
    style_function = style_function
).add_to(m)

m.save('index.html')
```
