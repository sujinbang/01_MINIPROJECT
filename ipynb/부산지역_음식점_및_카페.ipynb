```python
from collections import Counter 
file_path = './data/busan_food/6260000_부산광역시_07_24_03_P_외국인전용유흥음식점업.csv'

with open(file_path, encoding='euc-kr')as f :
    lines=f.readlines()
print(len(lines))

len_list=[]
for _ in range(len(lines)):
    print(str(_)+':'+str(len(lines[_].split(','))))
    len_list.append(len(lines[_].split(',')))
    
Counter(len_list)
```

```python
import csv
import pandas as pd

f = open(file_path, encoding='euc-kr')
reader = csv.reader(f)
csv_list=[]
for l in reader :
    csv_list.append(l)
f.close()
busan_foreign_food = pd.DataFrame(csv_list)
busan_foreign_food
```

```python
# 컬럼명 변경 
busan_foreign_food=busan_foreign_food.rename(columns=busan_foreign_food.iloc[0])
busan_foreign_food =busan_foreign_food.drop(busan_foreign_food.index[0])
busan_foreign_food
```

```python
# 전체 데이터에서 필요한 컬럼만 추출 
busan_foreign_food = busan_foreign_food[['업태구분명','인허가일자','폐업일자','영업상태명','사업장명','소재지전체주소','좌표정보(x)','좌표정보(y)']]
busan_foreign_food
```

```python
busan_foreign_food=busan_foreign_food[busan_foreign_food['영업상태명']=='영업/정상']
busan_foreign_food
```

```python
busan_total = pd.DataFrame
busan_total= pd.concat([busan_foreign_food,busan_normal_food,busan_snack_food])
```

```python
busan_total.info()
```

```python
# 주소 결측치 처리 
ex = busan_total[busan_total['소재지전체주소']==''].index
busan_total_ex = busan_total.drop(ex)
busan_total_ex
```

```python
busan_total_ex = busan_total_ex.reset_index(drop=True)
```

```python
busan_total_ex = busan_total_ex[['소재지전체주소','사업장명']]
```

```python
# 구 컬럼 생성 
list_addr_new= []
list_addr = busan_total_ex['소재지전체주소']
for i in list_addr.index :
#     print(list_addr[i][6:9])
    list_addr_new.append(list_addr[i][6:9])
```

```python
busan_total_ex['소재지주소(구)'] = list_addr_new
busan_total_ex
```

```python
busan_total_ex.to_csv('./data/busan_food/busan_total_ex.csv',index=False)
```
