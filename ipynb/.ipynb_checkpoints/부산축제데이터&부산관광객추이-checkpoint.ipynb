## 코로나 전/후 데이터 합치기

```python
df_aft_loc=df_aft_total.groupby('군/구')['총계'].mean()/2
df_aft_loc_desc=df_aft_loc.sort_values(ascending=False)
df_alt_loc_desc=round(df_aft_loc_desc)
df_bef_loc_desc=round(df_bef_loc_desc)
```

```python
df_aft_total=pd.DataFrame(df_aft_loc_desc)
df_bef_total=pd.DataFrame(df_bef_loc_desc)
```

```python
df_1=pd.merge(df_bef_total,df_aft_total,on='군/구')
```

```python
col_name=['2017년~2019년','2020년~2021년']
df_1.columns=col_name
df_1
```

```python
df_1['2017년~2019년']=df_1['2017년~2019년'].astype(int)
df_1['2020년~2021년']=df_1['2020년~2021년'].astype(int)
```

```python
df_1['코로나 전후']=df_1['2017년~2019년']-df_1['2020년~2021년']
```

```python
df_1=round(df_1)
df_1['감소율']=df_1['코로나 전후']/df_1['2017년~2019년']*100
df_1
```
