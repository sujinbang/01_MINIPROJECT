# 축제 파일 불러오기

```python
df_festival = pd.read_csv('./files/부산시_축제통합.csv')
df_festival = df_festival.drop('Unnamed: 0', axis=1)
df_festival_filt = df_festival[['축제명', '개최장소','위도','경도','축제시작_월']]
df_festival_filt
```

# 데이터 시각화

```python
file_path = "./maps/부산_시군구.geojson"
busan_sgg_get = json.load(open(file_path, encoding="utf-8"))
busan_sgg_get


print(len(busan_sgg_get['features']))
      
for i in range (len(busan_sgg_get['features'])):
    result = busan_sgg_get['features'][i]['properties']
    print(result)

m = folium.Map(
    location=[35.159144, 129.048293],
    #tiles = 'Stamen Terrain',
    zoom_start=10
    
)
    
tooltip = 'Click!'


#지도에 축제 표시
for i in df_festival_filt.index:
    
    folium.Marker([df_festival_filt.loc[i, "위도"], df_festival_filt.loc[i, "경도"]],
                popup=f'<div style="width:200px"><strong>{df_festival_filt.iloc[i]["축제명"]}</strong><br>\
                개최장소 : {df_festival_filt.iloc[i]["개최장소"]}<br>\
                축제시작(월) : {df_festival_filt.iloc[i]["축제시작_월"]}<br>\
                <img width="200px" src = ./image/festival.jpg><br>\
                <a href="http://www.bfo.or.kr/main/index.asp">부산문화관광축제조직위원회</a></div>',
                icon = folium.Icon(color='blue', icon='ok-circle'),
                
                tooltip = tooltip).add_to(m) 


# style 지정
def style_function(feature) :
    df = {
        "opacity" : 0.5,
        "weight" : 0.6,
        "color" : "yellow",
        "fillOpacity" : 0.5
    }

    return df
    
folium.GeoJson(
    busan_sgg_get,
    style_function = style_function
).add_to(m)

m.save('index2.html')
```
