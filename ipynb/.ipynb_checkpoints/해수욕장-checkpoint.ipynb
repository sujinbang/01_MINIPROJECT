# 해수욕장

```python
import pandas as pd
df = pd.read_excel('./files/원본/해수욕장_20220623155247.xlsx')
df = df.drop(0)
df = df.drop(1)
df = df.reset_index(drop=True)
df
```

```python
df1 = df.transpose()
df1.drop(df1.index[0])
df1 = df1.rename(columns=df1.iloc[0])
df1 = df1.drop(df1.index[0])
df1
```

```python
# x축에 넣기
list=[]
list.append(df1.index)
list
```

```python
for i in range(len(df1)):
    print(df1.index[i])
len(df1)
```

```python
import matplotlib.pyplot as plt
# 그래프 크기 지정하기 = 너비, 높이
plt.rcParams['figure.figsize'] = (16, 8)
# 선의 두께 지정
plt.rcParams['lines.linewidth'] = 2
# 폰트지정
plt.rc("font", family = "Malgun Gothic", size=10)

# 데이터 넣기
df1.plot()

plt.title("해수욕장")
plt.ylabel("방문객수")

# x축 범위?
#plt.xticks(list)
plt.xticks()

# 범례 위치 변경하기
plt.legend(loc="upper right")
# 그래프 내에 그리드선 표시하기
plt.grid(True)

plt.show()
```
