## 부산관광객 수 데이터
### 시계열데이터

### 한국관광공사 데이터를 활용해서 부산관광객수 데이터 시각화

## 데이터 불러오기

```python
import pandas as pd
```

```python
year=['2018','2019','2020','2021']
```

```python
visit_year=pd.DataFrame()
for i in year:
    file_path='./data/부산방문객수/{}_방문자 수.csv'.format(i)
    df=pd.read_csv(file_path,encoding='euc-kr')
    visit_year=pd.concat([visit_year,df])
```

```python
visit_year.columns=['기준연월','방문객수']
```

```python
visit_year=visit_year.reset_index()
```

## 데이터 결측치 확인

```python
k = []
for x in range(len(visit_year)):
    k.append(visit_year['기준연월'][x])
k
```

# 데이터 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 그래프 크키 저장하기 = 너비 높이

plt.rcParams['figure.figsize']=(20,10)
#선의 두꼐 지정
plt.rcParams['lines.linewidth']=2
#폰트저장
plt.rc('font',family ='Malgun Gothic',size=10)
#데이터넣기

visit_year.plot(color='red')
plt.title('2018년~2021년월별 부산 관광객수')
plt.ylabel('방문관광객수')
plt.ylim(8000000, 15000000) 

plt.xticks([i for i in range(0,48,1)], k, rotation = 55)

#범례위치 변경하기
plt.ticklabel_format(axis='y',style='plain')
plt.grid(True)
plt.legend(['방문자수'])
plt.show()
```
